```python
#Traduciendo filas de nuestro set de datos
df_habits['genero'] = df_habits['genero'].replace({
    'Male':'Masculino',
    'Female':'Femenino',
    'Other':'Otro'
})

df_habits["carrera"] = df_habits["carrera"].replace({
    "Arts": "Artes",
    "Psychology": "Psicologia",
    "Computer Science": "Ciencias de la Computacion",
    "Business": "Administracion",
    "Engineering": "Ingenieria",
    "Biology": "Biologia"
})

df_habits['trabajo_medio_tiempo'] = df_habits['trabajo_medio_tiempo'].replace({
    'Yes':'Si',
})

df_habits['riesgo_desercion'] = df_habits['riesgo_desercion'].replace({
    'Yes':'Si',
})

df_habits['ambiente_estudio'] = df_habits["ambiente_estudio"].replace({
    "Co-Learning Group": "Grupo de estudio colaborativo",
    "Library": "Biblioteca",
    "Quiet Room": "Sala silenciosa",
    "Dorm": "Dormitorio",
    "Cafe": "Cafetería"
})

df_habits['acceso_tutoria'] = df_habits['acceso_tutoria'].replace({
    'Yes':'Si',
})

df_habits['actividad_extracurricular'] = df_habits['actividad_extracurricular'].replace({
    'Yes':'Si',
})

df_habits["ingreso_familiar"] = df_habits["ingreso_familiar"].replace({
    "Low": "Bajo",
    "Medium": "Medio",
    "High": "Alto"
})

df_habits["estilo_aprendizaje"] = df_habits["estilo_aprendizaje"].replace({
    "Reading": "Lectura",
    "Kinesthetic": "Kinestésico",
    "Visual": "Visual",
    "Auditory": "Auditivo"
})

df_habits["calidad_dieta"] = df_habits["calidad_dieta"].replace({
    "Poor": "Mala",
    "Fair": "Aceptable",
    "Good": "Buena"
})

df_habits["calidad_internet"] = df_habits["calidad_internet"].replace({
    "Poor": "Mala",
    "Moderate": "Media",
    "High": "Buena"
})

df_habits["educacion_padres"] = df_habits["educacion_padres"].replace({
    "Some College": "Algunos estudios universitarios",
    "PhD": "Doctorado",
    "High School": "Escuela secundaria",
    "Master": "Maestría",
    "Bachelor": "Licenciatura"
})

df_habits.select_dtypes(include=['object'])
```

Se observa que los estudiantes que dedican más horas al estudio diario tienden a obtener mejores puntajes en los exámenes. A partir de las 4 horas diarias, se nota una mejora en la mediana del rendimiento y una disminución en la dispersión de los datos. Aunque la correlación es moderada, este patrón sugiere que una mayor dedicación al estudio puede influir positivamente en el desempeño académico.

```python
#Un entorno de estudio adecuado mejora el desempeño académico.

plt.figure(figsize=(10, 5))
sns.boxplot(data=df_habits, x='ambiente_estudio', y='puntaje_examen', hue='ambiente_estudio', palette='Set2', legend=False)
plt.title('Distribución del puntaje de examen según ambiente de estudio')
plt.xlabel('Ambiente de estudio')
plt.ylabel('puntaje_examen')
plt.show()
```

El gráfico muestra diferencias claras en el rendimiento según el lugar de estudio. Los estudiantes que estudian en salas silenciosas o en bibliotecas presentan las medianas de puntaje más altas y menor dispersión, lo que indica un entorno propicio para la concentración. En cambio, aquellos que estudian en el dormitorio tienen el rendimiento más bajo, con una mediana más baja y mayor variabilidad, lo que sugiere que este ambiente podría estar asociado a distracciones o menor efectividad. Esto refuerza la idea de que un entorno estructurado y silencioso favorece el desempeño académico.

```python
#Los estudiantes que duermen mas horas por día tienen mejores calificaciones.
def clasificar_sueño(horas):
  if horas < 7:
    return 'poco'
  elif  7 <= horas <= 9:
    return 'normal'
  elif horas > 9:
    return 'mucho'

df_habits['categoria_sueno'] = df_habits['horas_sueno'].apply(clasificar_sueño)

agrupar_sleep = df_habits.groupby('categoria_sueno')[['puntaje_examen']].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.boxplot(data=df_habits, x='categoria_sueno', y='puntaje_examen', hue='categoria_sueno', palette='Set2', legend=False)
plt.title('Distribución del puntaje de examen según categoría de sueño')
plt.xlabel('Categoría de Sueño')
plt.ylabel('puntaje_examen')
plt.show()
```

Los datos indican que los estudiantes que duermen entre 7 y 8 horas por noche alcanzan mejores resultados en los exámenes. La mediana del puntaje aumenta con la cantidad de horas de sueño, y la variabilidad disminuye. Este hallazgo resalta la importancia de un descanso adecuado para el rendimiento cognitivo y académico.

```python
#Estudiantes con mejor salud mental rinden mejor académicamente.

sns.boxplot(data=df_habits, x='salud_mental', y='puntaje_examen')
plt.title('Puntaje de examen según grupo de Salud Mental')
plt.xlabel('Grupo de Salud Mental')
plt.ylabel('puntaje_examen')
plt.show()
```

El gráfico muestra que los estudiantes con mejor salud mental (grupos más altos) tienden a obtener mayores puntajes en los exámenes. A medida que mejora el bienestar mental, aumenta la mediana del rendimiento y disminuye la dispersión de los resultados. Aunque la correlación es débil, el patrón es claro: un buen estado de salud mental parece ser un factor protector del rendimiento académico.

```python
#Estudiantes con menor nivel de estres rinden mejor academicamente
sns.boxplot(data=df_habits, x='nivel_estres', y='puntaje_examen')
plt.title('puntaje de examen según grupo de nivel de estres')
plt.xlabel('Grupo de Nivel de estres')
plt.ylabel('puntaje_examen')
plt.show()
```

A mayor nivel de estrés, menor rendimiento en el examen. # 🧠 Selección de variables relevantes

```python
#Elegimos variables
variables_utiles = [
    'nivel_motivacion',
    'nivel_estres',
    'salud_mental',
    'puntaje_examen',
    'promedio_anterior',
]
#Hacemos un df nuevo acortado
df_modelo = df_habits[variables_utiles + ['riesgo_desercion']].copy()
#Pasamos los si y no a 1 y 0
df_modelo['riesgo_desercion'] = df_modelo['riesgo_desercion'].map({'Si': 1, 'No': 0})
```

```python
# Importamos las herramientas para escalado y selección de características
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif

# Separamos las variables predictoras (X1) de la variable objetivo (y1)
X1 = df_modelo.drop("riesgo_desercion", axis=1)
y1 = df_modelo["riesgo_desercion"]

# Escalamos las variables predictoras para normalizarlas
scaler1 = StandardScaler()
X1_scaled = scaler1.fit_transform(X1)

# Seleccionamos las 3 variables más relevantes según ANOVA F-test
selector1 = SelectKBest(score_func=f_classif, k=3)
X1_selected = selector1.fit_transform(X1_scaled, y1)

# Obtenemos los nombres de las variables seleccionadas
selected_mask1 = selector1.get_support()
selected_features1 = [f for f, s in zip(X1.columns, selected_mask1) if s]
print("Variables seleccionadas desde df_modelo:", selected_features1)
```

### 🧠 Modelado y evaluación

Para abordar el problema de predicción del riesgo de deserción, se implementaron y compararon dos modelos supervisados de clasificación:


#### 📍 1. Regresión Logístic

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Definir variables y target
X = df_modelo[selected_features1]
y = df_modelo["riesgo_desercion"]

# Dividimos en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenamos el modelo
modelo_logit = LogisticRegression()
modelo_logit.fit(X_train, y_train)

# Predicción y evaluación
y_pred = modelo_logit.predict(X_test)


print("📊 LogisticRegression - Reporte de Clasificación:")
print(classification_report(y_test, y_pred))
print("📉 Matriz de Confusión:")
print(confusion_matrix(y_test, y_pred))
```

#### 🌳 2. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# 1. Variables seleccionadas
X = df_modelo[["nivel_motivacion", "nivel_estres", "promedio_anterior"]]
y = df_modelo["riesgo_desercion"]

# 2. Escalado
preprocessor = ColumnTransformer(
    transformers=[("num", StandardScaler(), X.columns)],
    remainder="passthrough"
)

# 3. Pipeline con Random Forest
rf_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier(n_estimators=100, random_state=42))
])

# 4. Separación de datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 5. Entrenamiento
rf_model.fit(X_train, y_train)


# 6. Predicción y métricas
y_pred_rf = rf_model.predict(X_test)

print("📊 Random Forest - Reporte de Clasificación:")
print(classification_report(y_test, y_pred_rf))
print("📉 Matriz de Confusión:")
print(confusion_matrix(y_test, y_pred_rf))
```
